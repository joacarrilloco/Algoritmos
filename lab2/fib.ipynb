## Iterative Fibonacci Comparation

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def fib( x ):
	if x == 0 :
		return 1
	i = 1
	j = 1
	for x in range(x - 1):
		j = j + i
		i = j - i
	return j
```

```python
def fibTime( x ):
    t = time.clock()
    fib( x )
    return time.clock() - t
```

```python
N = np.arange(0,5000, 10)
plt.plot(N,[fibTime(i) for i in N])
plt.xlabel('N')
plt.ylabel('Time')
plt.show()
```
