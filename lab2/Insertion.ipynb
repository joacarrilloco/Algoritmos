# Insertion Sort

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
```

```python
def insertionSort( arr ):
    steps = 1
    for i in range( 1, len( arr )):
        x = arr[i]
        j = i-1
        while (j > -1) and (arr[j] > x): 
            arr[j+1] = arr[j]
            j = j - 1
            steps+=3;
        arr[j+1] = x;
        steps+=4;
    return steps
```

## Memory limit using itertools until 12!

```python
def AverageTimeOfAllPermutations( N ):
    perms = list(itertools.permutations(np.arange( 1 , N + 1)))
    avg = 0
    for it in perms:
        avg = avg + insertionSort( list(it) )
    avg = avg/len(perms)
    return avg
```

Notice the Cuadratic aproximation of the complexity

```python
N = np.arange(1,12)
plt.plot(N,[AverageTimeOfAllPermutations(i) for i in N],'--k')
plt.title("Average Time Of N! Permutations")
plt.xlabel('N')
plt.ylabel('Time')
plt.show()
```
